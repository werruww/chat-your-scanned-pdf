```python
from PIL import Image
import pytesseract
import fitz

def extract_text_with_ocr(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        pix = page.get_pixmap()
        img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
        text += pytesseract.image_to_string(img)
    return text
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, AutoModel
import faiss
import torch
import numpy as np
import PyPDF2
from pdf2image import convert_from_path
import pytesseract
import os

# Load the base model (for text generation)
base_model_name = "FreedomIntelligence/RAG-Instruct-Llama3-3B"  # Or your preferred model
base_tokenizer = AutoTokenizer.from_pretrained(base_model_name)

# Add torch.dtype and device_map parameters
base_model = AutoModelForCausalLM.from_pretrained(
    base_model_name,
    torch_dtype="auto",  # or torch.float16
    device_map="auto",  # or specify a device map
)

# Load the embedding model
embedding_model_name = "nomic-ai/nomic-embed-text-v1"  # Or your preferred model
embedding_tokenizer = AutoTokenizer.from_pretrained(embedding_model_name)
embedding_model = AutoModel.from_pretrained(embedding_model_name)

def extract_text_from_pdf(pdf_path):
    """Extracts text from a PDF file using OCR if necessary."""
    try:
        with open(pdf_path, 'rb') as pdf_file:
            pdf_reader = PyPDF2.PdfReader(pdf_file)
            num_pages = len(pdf_reader.pages)
            extracted_text = ""
            for page_num in range(num_pages):
                page = pdf_reader.pages[page_num]
                extracted_text += page.extract_text()
            if extracted_text.strip():
                return extracted_text
            else:
                print("PyPDF2 failed to extract text. Trying OCR...")
                poppler_path = "/usr/bin"  # Adjust if needed
                os.environ['POPPLER_PATH'] = poppler_path
                images = convert_from_path(pdf_path, poppler_path=poppler_path)
                text = ""
                for image in images:
                    text += pytesseract.image_to_string(image)
                return text
    except Exception as e:
        print(f"Error during text extraction: {e}")
        return ""

def chunk_text(text, chunk_size=100):
    chunks = [text[i:i + chunk_size] for i in range(0, len(text), chunk_size)]
    return chunks

def generate_embeddings(text_chunks):
    embeddings = []
    for chunk in text_chunks:
        if chunk.strip():
            inputs = embedding_tokenizer(chunk, return_tensors="pt", truncation=True, padding=True)
            with torch.no_grad():
                outputs = embedding_model(**inputs)
            embeddings.append(outputs.last_hidden_state.mean(dim=1).cpu().numpy())
        else:
            print("Warning: Skipping empty chunk during embedding generation.")
    return embeddings

def create_faiss_index(embeddings):
    if not embeddings:
        raise ValueError("No embeddings to index.")
    index = faiss.IndexFlatL2(embeddings[0].shape[1])
    index.add(np.vstack(embeddings))
    return index

def search_chunks(query, index, top_k=5):
    query_embedding = generate_embeddings([query])[0]
    D, I = index.search(query_embedding, top_k)
    return I[0]

def run_chatbot(pdf_path):
    """Runs the chatbot, incorporating PDF content."""
    text = extract_text_from_pdf(pdf_path)
    if not text or not text.strip():
        print("Error: No text extracted from the PDF.")
        return
    text_chunks = chunk_text(text)
    embeddings = generate_embeddings(text_chunks)
    if not embeddings:
        print("Error: No embeddings generated.")
        return
    index = create_faiss_index(embeddings)
    while True:
        user_input = input("أنت: ")
        if user_input.lower() == "quit":
            break
        relevant_chunk_indices = search_chunks(user_input, index)
        valid_chunk_indices = [i for i in relevant_chunk_indices if 0 <= i < len(text_chunks)]
        if len(valid_chunk_indices) < len(relevant_chunk_indices):
            print("Warning: Some retrieved chunk indices are out of range and were ignored.")
        relevant_chunks = [text_chunks[i] for i in valid_chunk_indices]
        context = " ".join(relevant_chunks) + " " + user_input
        inputs = base_tokenizer.encode(context, return_tensors='pt')
        outputs = base_model.generate(inputs, max_length=300, num_return_sequences=1, do_sample=True, temperature=0.7)
        response = base_tokenizer.decode(outputs[0], skip_special_tokens=True)

        print("الروبوت:", response)

# Example usage
pdf_path = "/content/01_The_Lightning_Thief_1-19_1-1.pdf"
run_chatbot(pdf_path)
```
